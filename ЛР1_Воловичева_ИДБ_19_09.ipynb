```python
import numpy as np
import matplotlib.pyplot as plt

import json
import csv
```

Часть 1. Задача №1

```python
headers = ['item', 'country', 'year', 'sales']

with open('sales.json', 'r') as jsonfile:
    datajson = json.load(jsonfile)
    
with open('sales.csv', 'w') as csvfile: 
    writer = csv.DictWriter(csvfile, fieldnames = headers) 
    writer.writeheader() 
    
    for item in datajson:
        countrys = item['sales_by_country']
        
        for country in countrys:
            years = countrys[country]
            
            for year in years: 
                writer.writerow({'item':item['item'],
                                 'country':country,
                                 'year':year,
                                 'sales':item['sales_by_country'][country][year]})
```

```python
#Написать функцию transformation_plot, принимающую на вход набор двумерных точек и квадратную матрицу размером 2x2
#Отрисовать на одном графике оригинальные точки, на втором — точки после преобразования при помощи матрицы
#Отобразить точки таким образом, чтобы было понятно, какая из точек на первом графике соответствует какой точке на втором
def transformation_plot (dots, matrix):
    fig, a = plt.subplots(2)
    
    setting = dots[:,0] * dots[:,1]
    a[0].scatter(dots[:,0], dots[:,1], 70, setting) #scatter — график разброса
    
    mult = np.dot(dots,matrix) #функция dot вычисляет скалярное произведение двух массивов
    a[1].scatter(mult[:,0], mult[:,1], 70, setting)

colors = np.random.rand(50, 2)
position = np.array([[15,2],[15,1]])
transformation_plot(colors, position)
plt.show()
```

Задание №6

```python
#Дано множество векторов V размерности d. Дан вектор q такой же размерности. Определить:
#Пропорцию векторов v в V, для которых угол(v,q) < 90 градусов
#Пропорцию векторов v в V, для которых угол(v,q) < 30 градусов
v = np.random.rand(50, 2)
q = np.array([3, 1])
razm_1 = razm_2 = 0

for i in v:
    cos = i.dot(q)/(np.linalg.norm(q, ord=None, axis=None, keepdims=False)*np.linalg.norm(i, ord=None, axis=None, keepdims=False))
    #np.linalg.norm — данная функция определяет норму матрицы
    if (np.arccos(cos)*180/np.pi) < 90:
        razm_1 = razm_1 + 1
    if (np.arccos(cos)*180/np.pi) < 30:
        razm_2 = razm_2 + 1
        
razm_1 = razm_1/100
razm_2 = razm_2/100

print ("Пропорция векторов v в V, для которых угол (v,q) < 90 градусов:",razm_1)
print ("Пропорция векторов v в V, для которых угол (v,q) < 30 градусов:",razm_2)
```
